```python
import matplotlib.pyplot as plt
import numpy as np

# Data organization
datasets = ['Cleveland', 'Hungarian', 'Heart']

# Accuracies for different models
accuracies = {
    'SVM': [86.89, 77.97, 83.70],
    'RF': [86.89, 81.36, 88.04],
    'Hybrid(SVM/RF)': [86.89, 82.65, 86.41],
    'KNN': [91.80, 77.97, 83.70],
    'Logistic': [88.52, 83.05, 85.33],
    'Hybrid(KNN/Logistics)': [88.52, 81.36, 85.87],
    'Navy': [90, 75, 38],
    'Decision': [83, 69, 67],
    'Hybrid (Navy/Decision)': [88, 80, 68]
}

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

for i, dataset in enumerate(datasets):
    models = list(accuracies.keys())
    values = [accuracies[model][i] for model in models]
    
    ax[i].bar(models, values, color=['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'navy', 'purple', 'orange'])
    ax[i].set_title(f'Accuracies for {dataset} dataset')
    ax[i].set_ylabel('Accuracy (%)')
    ax[i].set_ylim(0, 100)
    ax[i].tick_params(axis='x', rotation=45)
    for index, value in enumerate(values):
        ax[i].text(index, value + 1, f'{value:.2f}%', ha='center')

plt.tight_layout()
plt.show()
```
